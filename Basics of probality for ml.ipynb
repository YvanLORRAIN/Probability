```python
import csv
import numpy as nu
from scipy import stats
import tkinter
from tkinter import *  
from tkinter import filedialog
import matplotlib.pyplot as plt
import os

class Q2:
    def __init__(self):
        # Parent Directory path 
        self.parent_dir ="D:/"
        self.Name="Year"
        self.year=[]
        self.yearCount=[]
        self.sample=[]
        #self.parent_dir+'/'+self.Name+'.csv'
        
    def openFile(self):
        # Open the .txt file and read it 
        with open(os.path.join(sys.path[0],"Year.csv"), newline='', encoding='utf8') as csvfile:
            reader = csv.reader(csvfile, delimiter=';')
            header = next(reader)

            # Insert data ito arrays
            if header != None:
                # Append years
                for row in reader:
                    if row[0]!="":
                        self.year.append(int(row[0]))
                    if row[1]!="":
                        self.sample.append(int(row[1]))
                    
            
q2=Q2()
q2.openFile()

# Compute the number of meteorites for each year
for i in range(len(q2.year)):
    c=0
    for j in range(len(q2.sample)):
        if q2.sample[j]==q2.year[i]:
            c=c+1
            #print(q2.year[i])
    q2.yearCount.append(c)
    #print(q2.year[i])
    #print(q2.year[i]," : ",q2.yearCount[i])
    
slope, intercept, r_value, p_value, std_err = stats.linregress(q2.year, q2.yearCount)

# plotting the points   
plt.scatter(q2.year, q2.yearCount)  
    
# naming the x axis  
plt.xlabel('Year')  
# naming the y axis  
plt.ylabel('Number of meteorites fallen')  
    
# giving a title to my graph  
plt.title('Number of meteorites fallen per year')     

# function to show the plot  
plt.show()  
```
